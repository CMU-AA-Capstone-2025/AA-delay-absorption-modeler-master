```python
import requests, pandas as pd
from io import StringIO

Hot_hubs = ['DFW', 'LAX', 'PHX', 'ORD', 'CLT', 'MIA', 'DCA', 'PHL', 'LGA', 'JFK']

weather_data = []

for tub in Hot_hubs:
  url = ("https://mesonet.agron.iastate.edu/cgi-bin/request/asos.py?"
        f"station={tub}&data=all&year1=2023&month1=01&day1=01&"
        "year2=2025&month2=01&day2=01&tz=Etc/UTC&format=onlycomma&"
        "latlon=no&direct=yes&report_type=1&report_type=2")

  r = requests.get(url, timeout=60)
  r.raise_for_status()

  df = pd.read_csv(StringIO(r.text))
  df["station_code"] = tub   # keep track of hub key

  df.to_csv(f"{tub}_weather_2023-01-01_to_2025-01-01.csv", index=False)

  weather_data.append(df)
  print(tub, len(df))

all_weather_data = pd.concat(weather_data, ignore_index=True)
all_weather_data.to_csv("All_Hubs_Weather_2023-01-01_to_2025-01-01.csv", index=False)

print("Combined dataset:", all_weather_data.shape)
```

```python

temps = []
visibilities = []

for metar_str in all_weather_data['metar']:
    m = Metar(metar_str)

    # Extract temperature in °C
    if m.temp is not None:
        temps.append(m.temp.value())

    # Extract visibility in meters
    if m.vis is not None:
        visibilities.append(m.vis.value())
```

```python
import numpy as np

def find_outliers_iqr(data):
    q1 = np.percentile(data, 25)
    q3 = np.percentile(data, 75)
    iqr = q3 - q1
    lower = q1 - 1.5 * iqr
    upper = q3 + 1.5 * iqr
    return [x for x in data if x < lower or x > upper]

temp_outliers_iqr = find_outliers_iqr(temps)
vis_outliers_iqr = find_outliers_iqr(visibilities)

print("Temperature outliers (IQR):", temp_outliers_iqr)
print("Visibility outliers (IQR):", vis_outliers_iqr)
```
